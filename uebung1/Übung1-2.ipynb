```python
import pandas as pd
import numpy as np
import urllib
import zipfile
import json
```

## Download and read data

### Download zip-file from internet

```python
urllib.urlretrieve ("http://ashleyw.co.uk/files/foods-2011-10-03.json.zip", "food.zip")
dateDownloaded = !date
dateDownloaded
```

### Extract json-file from zip-file

```python
fh = open('food.zip', 'rb')
z = zipfile.ZipFile(fh)
jsonfile = z.namelist()[0]
z.extract(jsonfile, "food/")
fh.close()
```

### Load json-file into variable

```python
fh = open('food/' + jsonfile, 'rb')
data = json.load(fh)
```

```python
print size(data)
data[0]['nutrients']
```

## Prepare data for analysis

### Load data into pandas dataFrame and drop nutrients information

```python
info = pd.DataFrame(data)
info = info.drop('nutrients',1)
info = info.drop('portions',1)
info = info.drop('tags',1)
```

### Load nutrients information in array of dataFrames

```python
nutrientArray = [pd.DataFrame for i in range(size(data))]
for i in range(size(data)):
    nutrientArray[i] = pd.DataFrame(data[i]['nutrients'])
    nutrientArray[i]['id'] = info['id'][i]
```

```python
nutrientArray[1]
```

```python
info.head()
```

### Concatenate nutrients array to form one single dataFrame

```python
nutrients = pd.concat(nutrientArray)
print size(nutrients)
nutrients.head()
```

### Remove duplicates

```python
nutrients = nutrients.drop_duplicates()
size(nutrients)
```

### Rename columns to produce unique names

```python
nutrients = nutrients.rename(columns={'description': 'nutrient'})
nutrients = nutrients.rename(columns={'group': 'nutGroup'})
nutrients.head()
```

### Merge remaining dataFrames (products and nutrients)

```python
products = pd.merge(info, nutrients, on='id', how='inner')
products.head()
```

## Analysis

### Create dataFrame for Zinc

```python
zinc = products[products['nutrient'] == 'Zinc, Zn']
zinc.head(10)
```

### Statistics for Zinc

```python
zinc.describe()
```
